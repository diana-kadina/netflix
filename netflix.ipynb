```python
import pandas as pd
#import matplotlib.pyplot as plt
netflix = pd.read_csv('netflix_data.csv')
netflix
```

```python
netflix_unique = netflix['type'].unique()
netflix_unique
```

```python
netflix_subset = netflix[netflix['type'] == 'Movie'] 
netflix_subset.head(10)
netflix_movies = netflix_subset[['title', 'country', 'genre', 'release_year', 'duration']]
netflix_movies
```

```python
import matplotlib.pyplot as plt
#shorter than 60 minutes
short_movies = netflix_movies.loc[netflix_movies['duration'] < 60]
short_movies
colors = []
for genre in netflix_movies['genre']:
    if genre == 'Children':
        colors.append('green')
    elif genre == 'Stand-Up':
        colors.append('yellow')
    elif genre == 'Documentaries':
        colors.append('blue')
    else:
        colors.append('grey')

fig = plt.figure(figsize=(12,8))
plt.scatter(netflix_subset['release_year'], netflix_subset['duration'], c=colors)
plt.xlabel('Release year')
plt.ylabel('Duration (min)')
plt.title('Movie Duration by Year of Release')
plt.show()
```
